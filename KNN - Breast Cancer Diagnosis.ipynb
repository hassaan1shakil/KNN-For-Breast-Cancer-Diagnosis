```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("KNNAlgorithmDataset.csv")
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

duplicates = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicates)

summary = df.describe()
print("\nSummary Statistics:\n", summary)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop('diagnosis', axis=1)
Y = df['diagnosis']
X_train, X_test,  Y_train, Y_test = train_test_split(X, Y, train_size=0.6, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
def KNN(X_train, X_test, k=5):

    results = list()
    classes_map = dict()
    predictions = list()

    for test_point in X_test:

      for i in range(len(X_train)):
        distance = np.linalg.norm(X_train[i] - test_point)
        results.append((Y[i], distance))

        results.sort(key = lambda x: x[1])

      for j in range(k):
        if results[j][0] in classes_map:
          classes_map[results[j][0]] += 1

        else:
          classes_map[results[j][0]] = 1


      sorted_map = sorted(classes_map.items(), key = lambda x: x[1], reverse=True)
      first_key = next(iter(sorted_map))
      predictions.append(first_key[0])

#       if i % 10 == 0:
#           print(f"Predicted Class of Test Point: {first_key[0]}")

      results.clear()
      classes_map.clear()
    
    return predictions
```

```python
# Finding the Best K

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

k = 1
max_accuracy = 0

for i in range(1, 32, 2):
    predictions = KNN(X_train, X_test, i)
    accuracy = accuracy_score(Y_test, predictions)
    if accuracy > max_accuracy:
        max_accuracy = accuracy
        k = i

print(f"Max Accuracy of {max_accuracy} achieved by using K = {i}")
```

```python
# Mesuring Accuracy of KNN using the Best K

predictions = KNN(X_train, X_test, k)

accuracy = accuracy_score(Y_test, predictions)
precision = precision_score(Y_test, predictions, pos_label='M')
recall = recall_score(Y_test, predictions, pos_label='M')
f1 = f1_score(Y_test, predictions, pos_label='M')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
